```python
import tensorflow as tf
```

```python
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.optimizers import Adam
from keras.utils import to_categorical
import keras
import keras.backend as tf

import numpy as np
```

```python
%pylab inline
```

```python
batch_size = 32
num_classes = 10
epochs = 100
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
fig, ax = pylab.subplots(3, 5, figsize=(12, 8))

for i in range(15):
    ax[i % 3, i % 5].set_xlabel(y_train[i])
    ax[i % 3, i % 5].imshow(x_train[i,:,:,:])
```

```python
x_train = x_train.astype('float') / 255.0
x_test = x_test.astype('float') / 255.0
print(x_train.min(), x_train.max())
print(x_test.min(), x_test.max())
```

```python
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
print('y_train = {}'.format(y_train[0]))
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=x_train.shape[1:],
                 padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu', padding='valid'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss = keras.losses.categorical_crossentropy,
              optimizer = Adam(),
              metrics = ['accuracy'])
```

```python
model.summary()
```

```python

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_split=0.3)

score = model.evaluate(x_test, y_test, verbose=0)
print('\n\nTest loss:', score[0])
print('Test accuracy:', score[1])
```

```python
def plot_history(history):
    pylab.plot(history.history['accuracy'])
    pylab.plot(history.history['val_accuracy'])
    pylab.title('model accuracy')
    pylab.ylabel('accuracy')
    pylab.xlabel('epoch')
    pylab.legend(['train', 'val'], loc='upper left')
    pylab.show()
```

```python
plot_history(history)

score = model.evaluate(x_test, y_test, verbose=1)
print('\n\nTest loss:', score[0])
print('Test accuracy:', score[1])
```

```python
model_file = 'simple_model_cifar10.h5'
model.save(model_file)
```
